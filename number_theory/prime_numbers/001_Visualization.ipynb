```python
import numpy as np
import sympy as sp

import matplotlib.pyplot as plt

jtplot.style(figsize=(15, 9))
```

```python
# number of numbers
N = 1000
```

```python
prime = np.array(list(sp.primerange(0, N)))
prime
```

```python
def to_polar_cs(n):
    return np.array([x*np.cos(x) for x in n]), np.array([y*np.sin(y) for y in n])
```

```python
polar_x, polar_y = to_polar_cs(prime)
polar_x_full, polar_y_full = to_polar_cs(np.arange(0, N))
```

```python
fig, ax = plt.subplots(1, figsize=(11, 11))
ax.scatter(polar_x_full, polar_y_full, color='b', alpha=0.3)
ax.scatter(polar_x, polar_y, color='w')
```
