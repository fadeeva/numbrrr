# Box-Muller Transform

```python
import numpy as np
import scipy as sp

import matplotlib.pyplot as plt
```

```python
r = np.random.uniform(low=0.0, high=1.0, size=1_000)
f = np.random.uniform(low=0.0, high=1.0, size=1_000)
```

```python
z0 = np.cos(2*np.pi*f)*np.sqrt(-2*np.log(r))
z1 = np.sin(2*np.pi*f)*np.sqrt(-2*np.log(r))
```

```python
sp.stats.shapiro(z0), sp.stats.shapiro(z1)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(21, 9))

ax[0, 0].hist(r, bins=50)
ax[0, 0].set_title('r, Uniform Distribution')

ax[0, 1].hist(f, bins=50)
ax[0, 1].set_title('r, Uniform Distribution')

ax[1, 0].hist(z0, bins=50)
ax[1, 0].set_title('z0, Normal Distribution')

ax[1, 1].hist(z1, bins=50)
ax[1, 1].set_title('z1, Normal Distribution');
```
