# Visualization of Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Normal Distribution

```python
mean = 0
sigma = 1
x = np.linspace(-5, 5, 100)
k = (x - mean) / sigma
```

```python
fig, axs = plt.subplots(2, 2, figsize=(21, 13))

axs[0, 0].plot(x, np.exp(k))
axs[0, 0].set_title(r'$e^k$', fontsize=21)

axs[0, 1].plot(x, np.exp(k**2))
axs[0, 1].set_title(r'$e^{k^2}$', fontsize=21)

axs[1, 0].plot(x, np.exp(-k**2/2))
axs[1, 0].set_title(r'$e^{\frac{-k^2}{2}}$', fontsize=21)

axs[1, 1].plot(x, (1/(sigma*np.sqrt(2*np.pi)))*np.exp(-k**2/2))
axs[1, 1].set_title(r'$\frac{1}{\sigma\sqrt{2\pi}}e^{\frac{-k^2}{2}}$', fontsize=21)

fig.tight_layout(pad=5)
```

```python
fig, ax = plt.subplots(1, figsize=(11, 7))

sp1 = ax.plot(x, np.exp(-k**2/2), c='b', label=r'$e^{\frac{-k^2}{2}}$')
ax_ = ax.twiny()
sp2 = ax_.plot(x, (1/(sigma*np.sqrt(2*np.pi)))*np.exp(-k**2/2), c='g', label=r'$\frac{1}{\sigma\sqrt{2\pi}}e^{\frac{-k^2}{2}}$')

subs = sp1+sp2
subs_labels = [label.get_label() for label in subs]
ax.legend(subs, subs_labels, fontsize=21)
```

## Cauchy Distribution 

```python
# standard
x_0 = 0
gamma = 1
x = np.linspace(-5, 5, 100)
k = (x - x_0)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(21, 13))

axs[0, 0].plot(x, k**2)
axs[0, 0].set_title(r'$k^2$', fontsize=21)

axs[0, 1].plot(x[x<0], 1/x[x<0]**2, c='b')
axs[0, 1].plot(x[x>0], 1/x[x>0]**2, c='b')
axs[0, 1].set_title(r'$\frac{1}{k^2}$', fontsize=21)

axs[1, 0].plot(x, gamma/(k**2 + gamma))
axs[1, 0].set_title(r'$\frac{\gamma}{k^2 + \gamma}$', fontsize=21)

axs[1, 1].plot(x, (1/np.pi)*(gamma/(k**2 + gamma)))
axs[1, 1].set_title(r'$\frac{1}{\pi}[\frac{\gamma}{k^2 + \gamma}]$', fontsize=21)

fig.tight_layout(pad=5)
```

```python
fig, ax = plt.subplots(1, figsize=(11, 7))

sp1 = ax.plot(x, gamma/(k**2 + gamma), c='b', label=r'$\frac{\gamma}{k^2 + \gamma}$')
ax_ = ax.twiny()
sp2 = ax_.plot(x, (1/np.pi)*(gamma/(k**2 + gamma)), c='g', label=r'$\frac{1}{\pi}[\frac{\gamma}{k^2 + \gamma}]$')

subs = sp1+sp2
subs_labels = [label.get_label() for label in subs]
ax.legend(subs, subs_labels, fontsize=21)
```
