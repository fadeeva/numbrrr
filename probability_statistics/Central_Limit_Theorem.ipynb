# Central Limit Theorem

```python
import numpy as np

import scipy.stats as sps

import matplotlib.pyplot as plt
import seaborn as sns

jtplot.style(figsize=(15, 9))
```

```python
s = np.array([])
N = 1000
for _ in range(10000):
    sample = sps.maxwell.rvs(size=N, loc=1)
    sum_s = sample.sum()
    std_s = sample.std()
    result = (sum_s - N) / np.sqrt(N)*std_s
    s = np.append(s, result)
    
sps.normaltest(s).pvalue
```

```python
sns.histplot(s, kde=True, color='skyblue', bins=1000)
```
