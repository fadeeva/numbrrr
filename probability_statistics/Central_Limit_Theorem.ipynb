# Central Limit Theorem

```python
import numpy as np

import scipy.stats as sps

import matplotlib.pyplot as plt
import seaborn as sns

jtplot.style(figsize=(15, 9))
```

```python
s = np.array([])
for _ in range(100):
    s = np.append(s, sps.levy.rvs(size=1000).mean())
s[:10]
```

```python
sns.histplot(s, kde=True, color='skyblue')
```

```python
r = sps.levy.rvs(size=1000)
a, b = sps.levy.ppf(0), sps.levy.ppf(0.6)
x = np.linspace(a, b, 100)
fig, ax = plt.subplots(1, 1)

ax.plot(x, sps.levy.pdf(x), 'r-', lw=3, alpha=0.6, label='gamma pdf')
bins = np.concatenate((np.linspace(a, b, 20), [np.max(r)]))
ax.hist(r, density=True, bins=bins, histtype='stepfilled', alpha=0.5)
ax.set_xlim([x[0], x[-1]])
ax.legend(loc='best')
```

```python
a = 1.99
r = sps.gamma.rvs(a, size=1000)
x = np.linspace(sps.gamma.ppf(0.01, a),
                sps.gamma.ppf(0.99, a), 100)
fig, ax = plt.subplots(1, 1)

ax.plot(x, sps.gamma.pdf(x, a), 'r-', lw=3, alpha=0.6, label='gamma pdf')

ax.hist(r, density=True, bins='auto', histtype='stepfilled', alpha=0.5)
ax.set_xlim([x[0], x[-1]])
ax.legend(loc='best')
```
