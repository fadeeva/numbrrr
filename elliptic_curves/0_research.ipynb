```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y, x = np.ogrid[-5:5:100j, -5:5:100j]
```

```python
fig, axs = plt.subplots(1, 5, figsize=(21, 3))

axs[0].contour(x.ravel(), y.ravel(), y**2 - x**3 + 1, [0], colors='b')
axs[1].contour(x.ravel(), y.ravel(), y**2 - x**3 - 1, [0], colors='b')
axs[2].contour(x.ravel(), y.ravel(), y**2 - x**3 + 3*x - 3, [0], colors='b')
axs[3].contour(x.ravel(), y.ravel(), y**2 - x**3 + 4*x, [0], colors='b')
axs[4].contour(x.ravel(), y.ravel(), y**2 - x**3 + x, [0], colors='b')

axs[0].set_title(r'$y^2=x^3-1$')
axs[1].set_title(r'$y^2=x^3+1$')
axs[2].set_title(r'$y^2=x^3-3x+3$')
axs[3].set_title(r'$y^2=x^3-4x$')
axs[4].set_title(r'$y^2=x^3-x$')
```

```python
a, b = -1, 1
plt.contour(x.ravel(), y.ravel(), y**2 - x**3 - x * a - b, [0], colors='b')
plt.grid(True, color='w', alpha=.1)
```
